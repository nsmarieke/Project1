### This notebook contains the first observations of the data

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data and show first 5 rows (all columns)
df = pd.read_csv('Regularities_by_liaisons_Trains_France.csv')
pd.set_option("display.max_columns", None)
df.head()
```

```python
# Look at data
shape_df = df.shape
no_nulls = set(df.columns[df.isnull().mean()==0])
number_columns_no_null = len(no_nulls)
most_missing = set(df.columns[df.isnull().mean()>0.75])
number_most_missing = len(most_missing)
number_of_rows_nan = df[df.isnull().all(axis=1)].shape
years = df['Year'].unique()
months = df['Month'].unique()
period_2015 = df[df['Year'] == 2015]
period_2015 = period_2015['Month'].unique()
period_2020 = df[df['Year'] == 2020]
period_2020 = period_2020['Month'].unique()

print('Number of rows:', shape_df[0],', number of columns:', shape_df[1])
print('columns with no null values:', no_nulls, ', number of columns', number_columns_no_null)
print('columns who have more than 75% missing:', most_missing, ', number of columns', number_most_missing)
print('Number of rows with all NaN values:', number_of_rows_nan[0])
print('Unique years:', years)
print('Unique months:', months)
print('months of 2015:', period_2015)
print('months of 2020:', period_2020)
print('data types:', df.dtypes)
```

## Observations
### The dataframe contains only 8 columns of the 34 with no missing values
### There are two columns with more than 75% missing values
### There are no rows with all missing values
### The data contains 6 years of data from January 2015 until June 2020

```python
# Plots
plt.plot(df['Average delay of all departing trains (min)'])
plt.show()
plt.plot(df['Average delay of late departing trains (min)'])
plt.show()
```

## Observations
### First plot: Average delays far below zero
### Second plot: The highest peak is above 300 minutes
